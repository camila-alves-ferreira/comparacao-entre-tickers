# 0. Introdução

A análise abaixo foi feita utilizando dados do Yahoo Finance. A motivação é observar alguns comportamentos de dois dos tickets de maior participação do índice S&P 500, que representam uma participação de 14% do índice geral - MSFT (Microsoft) e AAPL (Apple).

Fonte de dados: https://finance.yahoo.com/

# 1. Instalação e Importação das Bibliotecas

Primeiramente utilizaremos a biblioteca yfinance para baixar os dados históricos:

```python
!pip install yfinance
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

Vamos carregar os dados da ação Microsoft.

```python
msft = yf.Ticker("MSFT")
```

Podemos observar os preços de fechamento com intervalo de uma semana ao longo do ano de 2023 para o ticker MSFT:

```python
msft_price = yf.download("MSFT", start="2023-01-01", end="2023-12-31", interval="1wk")["Close"]
msft_price
```

Os preços da ação podem ser visualizados de maneira mais clara em um dataframe:

```python
df_msft_price = pd.DataFrame(msft_price)
df_msft_price
```

```python
msft.info
```

O pagamento de dividendos dos últimos dez anos de MSFT pode ser observado no dataframe abaixo:

```python
msft_dividends = yf.download("MSFT", start="2013-01-01", end="2023-12-31", actions=True)['Dividends']
msft_dividends
```

```python
df_msft_dividends = pd.DataFrame(msft_dividends)
df_msft_dividends
```

Vamos carregar dados também para Apple (AAPL).

```python
aapl = yf.Ticker("AAPL")
```

Podemos observar os preços de fechamento com intervalo de uma semana ao longo do ano de 2023 para o ticker AAPL:

```python
aapl_price = yf.download("AAPL", start = "2023-01-01", end = "2023-12-31", interval= "1wk")["Close"]
aapl_price
```

Os preços da ação podem ser visualizados de maneira mais clara em um dataframe:

```python
df_aapl_price = pd.DataFrame(aapl_price)
df_aapl_price
```

Para obter a lista de informações do ticker deve-se usar o comando info:

```python
aapl.info
```

```python
aapl_dividends = yf.download("AAPL", start="2013-01-01", end="2023-12-31", actions=True)['Dividends']
aapl_dividends
```

O pagamento de dividendos dos últimos dez anos de AAPL pode ser observado no dataframe abaixo:

```python
df_aapl_dividends = pd.DataFrame(msft_dividends)
df_aapl_dividends
```

# Comparação entre os tickets

Através de gráficos podemos ter uma visualização e melhor compreensão dos dados. Nos gráficos abaixo são feitas comparações de parâmetros já analisados como preço de fechamento ao longo de 2023 e dividendos dos últimos dez anos.

```python
plt.plot(msft_price, label="MSFT")
plt.plot(aapl_price, label = "AAPL")
plt.legend()
plt.show()
```

```python
msft_dividends = yf.download("MSFT", start = "2013-01-01", end = "2023-12-31", actions=True)['Dividends']
msft_dividends
```

```python
aapl_dividends = yf.download("AAPL", start = "2013-01-01", end = "2023-12-31", actions=True)['Dividends']
aapl_dividends
```

A Microsoft frente à Apple se apresenta como uma melhor pagadora de dividendos num longo prazo.

```python
plt.plot(msft_dividends, label="MSFT", color="tomato")
plt.plot(aapl_dividends, label="AAPL", color="teal")
plt.title("Dividendos de MSFT e AAPL de 2013 a 2023")
plt.xlabel("Ano")
plt.ylabel("Dividendos em US$")
plt.legend()
plt.show()
```
